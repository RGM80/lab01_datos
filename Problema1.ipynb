```python
def calcular_probabilidad_triumfo(Ganados, Empatados, Perdidos):
    Jugados = Ganados + Empatados + Perdidos
    if Jugados == 0:
        return 0  # Evita división por cero si no hay partidos jugados
    return Ganados / Jugados 
```
